## Task 1

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    # strip and title case for readable names
    clean_name = student["name"].strip().title()
    # cast to int for math/logic ops
    clean_roll = int(student["roll"])
    # split and convert to list of ints
    clean_marks = [int(m) for m in student["marks_str"].split(", ")]
    
    # check if name only contains letters
    is_valid = all(word.isalpha() for word in clean_name.split())
    status = "✓ Valid name" if is_valid else "✗ Invalid name"
    
    # store for later access
    student_data = {"name": clean_name, "roll": clean_roll, "marks": clean_marks}
    cleaned_students.append(student_data)
    
    # simple f-string profile card
    print(f"================================\n"
          f"Student : {clean_name}\n"
          f"Roll No : {clean_roll}\n"
          f"Marks   : {clean_marks}\n"
          f"================================ {status}")

# target specific student by roll
s103 = next(s for s in cleaned_students if s["roll"] == 103)
print(f"\nStudent 103 Case Check:\nUPPER: {s103['name'].upper()}\nlower: {s103['name'].lower()}")
```

## Task 2

```python
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

# process existing grades
print(f"Report Card: {student_name}")
for sub, score in zip(subjects, marks):
    # range checks for grading
    if score >= 90: grade = "A+"
    elif score >= 80: grade = "A"
    elif score >= 70: grade = "B"
    elif score >= 60: grade = "C"
    else: grade = "F"
    print(f"{sub}: {score} ({grade})")

# basic stats 
total = sum(marks)
avg = round(total / len(marks), 2)
# find index of max/min to map back to subject name
high_idx = marks.index(max(marks))
low_idx = marks.index(min(marks))

print(f"\nTotal: {total} | Avg: {avg}")
print(f"Highest: {subjects[high_idx]} ({marks[high_idx]})")
print(f"Lowest: {subjects[low_idx]} ({marks[low_idx]})")

# dynamic entry system
new_count = 0
while True:
    sub_input = input("\nEnter subject (or 'exit' to quit): ").strip()
    if sub_input.lower() == 'exit':
        break
    
    try:
        # validate numeric input
        score_input = float(input(f"Enter marks for {sub_input}: "))
        if 0 <= score_input <= 100:
            subjects.append(sub_input)
            marks.append(score_input)
            new_count += 1
        else:
            print("*CAUTION*: do not add invalid entries to the list")
    except ValueError:
        # catch non-numeric strings
        print("*CAUTION*: do not crash")

# final stats update
updated_avg = round(sum(marks) / len(marks), 2)
print(f"\nAdded {new_count} new subject(s).")
print(f"Updated Average: {updated_avg}")
```

## Task 3

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

processed_data = []
pass_count = 0
fail_count = 0

print(f"{'Name':<18} | {'Average':<7} | Status")
print("-" * 40)

for name, marks in class_data:
    # calculate mean for each tuple
    avg = round(sum(marks) / len(marks), 2)
    # simple binary status check
    status = "Pass" if avg >= 60 else "Fail"
    
    # track counts for summary
    if status == "Pass": pass_count += 1 
    else: fail_count += 1
    
    processed_data.append({"name": name, "avg": avg})
    # aligned printing using f-string padding
    print(f"{name:<18} | {avg:<7.2f} | {status}")

# derive class stats from the processed list
all_averages = [s["avg"] for s in processed_data]
class_avg = round(sum(all_averages) / len(all_averages), 2)
# find student with highest average
topper = max(processed_data, key=lambda x: x["avg"])

print("-" * 40)
print(f"Passed: {pass_count} | Failed: {fail_count}")
print(f"Class Topper: {topper['name']} ({topper['avg']})")
print(f"Overall Class Average: {class_avg}")
```

## Task 4

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: initial cleanup to remove padding
clean_essay = essay.strip()
print(f"Cleaned Essay: {clean_essay}")

# Step 2: title case for headers/titles
print(f"Title Case: {clean_essay.title()}")

# Step 3: case-insensitive count (clean_essay is lowercase here)
python_count = clean_essay.count("python")
print(f"Python Count: {python_count}")

# Step 4: simple swap to add branding/emojis
branded_essay = clean_essay.replace("python", "Python 🐍")
print(f"Branded: {branded_essay}")

# Step 5: break into list based on sentence endings
sentences = clean_essay.split(". ")
print(f"Sentence List: {sentences}")

# Step 6: clean formatting and numbering
print("\nNumbered Sentences:")
for i, sentence in enumerate(sentences, 1):
    # strip to clean potential trailing spaces
    s = sentence.strip()
    # ensure every sentence ends with a period
    if not s.endswith("."):
        s += "."
    print(f"{i}. {s.capitalize()}")
```
